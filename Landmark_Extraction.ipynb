```
from scipy.spatial import distance as dist
from mlxtend.image import extract_face_landmarks
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```
from google.colab import drive
drive.mount('/content/drive')
```

```
folder_path = '/content/drive/MyDrive/Team Epic - Driver Drowsiness Detection'
features = []
individuals = []
labels = []
```

```
for i in range(13,16):
  for j in range(0,11,5):
    path = folder_path + '/' + str(i) + '_' + str(j) + '.npy'
    frames = np.load(path)
    plt.imshow(frames[10])
    plt.show()
   

```

```
for i in range(0,11,5):
  path = folder_path + '/' + str(13) + '_' + str(i) + '.npy'
  frames = np.load(path)
  print('Number of frames: ', len(frames))
  counter = 0
  # Extract landmarks
  for frame in frames:
    frame = cv2.rotate(frame, cv2.ROTATE_90_COUNTERCLOCKWISE)
    landmarks = extract_face_landmarks(frame)
    if (sum(landmarks)[0] != 0):
      features.append(landmarks)
      individuals.append(13)
      labels.append(i)
      counter += 1
  print('Number of frames with detectable faces: ', counter)
```

```
# 14TH individual alert video
path = folder_path + '/' + str(14) + '_' + str(0) + '.npy'
frames = np.load(path)
print('Number of frames: ', len(frames))
counter = 0
# Extract landmarks
for frame in frames:
  frame = cv2.rotate(frame, cv2.ROTATE_90_COUNTERCLOCKWISE)
  landmarks = extract_face_landmarks(frame)
  if (sum(landmarks)[0] != 0):
    features.append(landmarks)
    individuals.append(14)
    labels.append(0)
  counter += 1
print('Number of frames with detectable faces: ', counter)
```
